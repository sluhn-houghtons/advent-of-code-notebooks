## Example

```python
example = ['+1', '-2', '+3', '+1']

def calculate_frequency(input):
  arr = [int(delta) for delta in input]
  return sum(arr)

frequency = calculate_frequency(example)
frequency
```

## Part 1

```python
puzzle_input = open('./input.txt', 'r').read().split('\n')

calculate_frequency(puzzle_input)
```

# Part 2

```python
example = ['+3', '+3', '+4', '-2', '-4']

def calculate_frequency_loop(input):
  frequency_set = set([])
  current_frequency = 0
  while True:
    for delta in input:
      current_frequency += int(delta)
      if (current_frequency in frequency_set):
        return current_frequency
      frequency_set.add(current_frequency)

calculate_frequency_loop(example)
```

```python
calculate_frequency_loop(puzzle_input)
```
